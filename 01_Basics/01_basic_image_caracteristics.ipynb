```python
from skimage import data
import numpy as np
import matplotlib.pyplot as plt

# Image from data
img_camera = data.camera()
```

```python
print("Image: Camera")
print(img_camera)
```

```python
# 8x8 Noise image
height = 8
width = 8
img_noise = np.random.rand(height,width)
img_noise = (img_noise * 16).astype(np.uint8) # 16 color Quantization
print("Image: Noise")
print(img_noise)
```

```python
print("Img_camera:\n")
print(f"Dimension img_camera: {img_camera.ndim}")
print(f"Resolution img_camera: {img_camera.shape[1]} x {img_camera.shape[0]}")
print(f"Total img_camera size: {img_camera.size}")
print(f"Min value img_camera: {np.min(img_camera)}")
print(f"Max value img_camera: {np.max(img_camera)}")
print(f"Mean value img_camera: {np.mean(img_camera):.2f}")
print(f"Amount of particular pixel value img_camera: {np.sum(img_camera == 25)}")
# ([[1,2,3],[3,4,5]] == 3) === [[F,F,T],[T,F,F]] === [[0,0,1],[1,0,0]]
fig, ax = plt.subplots(1,2)
ax[0].axis("off")
ax[0].set_title("Image Camera")
ax[0].imshow(img_camera,cmap='grey')
ax[1].set_title("img Camera histogram")
ax[1].hist(np.ravel(img_camera), bins=256)
#ravel => [[1,2,3],[3,4,5]] -> [1,2,3,3,4,5] 
plt.tight_layout(pad=1, w_pad=2.5, h_pad=1.0)
plt.show()

print("\n---------------------------------------------------------------------------\n")
print("Img_noise:\n")
print(f"Dimension img_noise: {img_noise.ndim}")
print(f"Resolution img_noise: {img_noise.shape[1]} x {img_noise.shape[0]}")
print(f"Total img_noise size: {img_noise.size}")
print(f"Min value img_noise: {np.min(img_noise)}")
print(f"Max value img_noise: {np.max(img_noise)}")
print(f"Mean value img_noise: {np.mean(img_noise):.2f}")
print(f"Amount of particular pixel value img_noise: {np.sum(img_noise == 4)}")
fig, ax = plt.subplots(1,2)
ax[0].axis("off")
ax[0].set_title("Image Noise")
ax[0].imshow(img_noise,cmap='grey')
ax[1].set_title("img Camera histogram")
ax[1].hist(np.ravel(img_noise), bins=8)
plt.tight_layout(pad=1, w_pad=2.5, h_pad=1.0)
plt.show()

```
