# Indexation

```python
from skimage import data
import numpy as np
import matplotlib.pyplot as plt

# Image from data
img_camera = data.camera()

# 8x8 Noise image
height = 8
width = 8
img_noise = np.random.rand(height,width)
img_noise = (img_noise * 16).astype(np.uint8) # 16 color Quantization
```

Indexing a specific pixel

```python
print(f"Light value of img_camera in row 19, column 8: {img_camera[19, 8]}") # img_camera[19, 8] === img_camera[19][8]
```

Indexing with ranges

```python
# start: 0 | end: size | step: 1
# This is a reference to the original image

fig, ax = plt.subplots(2,2)
ax[0][0].axis("off")
ax[0][0].set_title("Original")
ax[0][0].imshow(img_camera, cmap='grey')

cut = img_camera[50:200, 100:400]
ax[0][1].axis("off")
ax[0][1].set_title("Cutted")
ax[0][1].imshow(cut, cmap='grey')

removing_half_x = img_camera[:, 20:400:2] #Range only on x and skiping 1 each
ax[1][0].axis("off")
ax[1][0].set_title("Cutting and skipping on X")
ax[1][0].imshow(removing_half_x, cmap='grey')

pixelated = img_camera[::10, ::10] #It will look the same on plt bc it recises but actually, less resolution
ax[1][1].axis("off")
ax[1][1].set_title("Pixelated")
ax[1][1].imshow(pixelated, cmap='grey')

plt.show()
```

Using the ranges without copy, will mess with the original photo

```python
range = img_camera[20:100, ::5]

#Modify on range view
range[0:50, :] = 0  #Change all pixels y:0,50 to black

fig, ax = plt.subplots(1,2)
ax[0].axis("off")
ax[0].set_title("Range view modified")
ax[0].imshow(range, cmap='grey')

ax[1].axis("off")
ax[1].set_title("Original")
ax[1].imshow(img_camera, cmap='grey')
```
